```python
import numpy as np
import pickle
import gzip
import os
from PIL import Image
import operator
```

```python
TrainingDataFile = 'mnist.pkl.gz'
Data = gzip.open(TrainingDataFile,'rb')
training,validation,testing= pickle.load(Data,encoding='latin1')
```

```python
trainingData = training[0]
trainingTarget = training[1]
validationData = validation[0]
validationTarget = validation[1]
testingData = testing[0]
testingTarget = testing[1]
USPSMat  = []
USPSTar  = []
curPath  = 'USPSdata/Numerals'
print(type(validationData))
```

```python
for j in range(0,10):
    curFolderPath = curPath + '/' + str(j)
    imgs =  os.listdir(curFolderPath)
    for img in imgs:
        curImg = curFolderPath + '/' + img
        if curImg[-3:] == 'png':
            img = Image.open(curImg,'r')
            img = img.resize((28, 28))
            savedImg = img
            imgdata = (255-np.array(img.getdata()))/255
            USPSMat.append(imgdata)
            USPSTar.append(j)
USPSMat = np.asarray(USPSMat)
USPSTar = np.asarray(USPSTar)
t = np.zeros(shape=(50000,10))
for x in range(len(trainingData)):
    t[x][trainingTarget[x]] = 1

v = np.zeros(shape=(10000,10))
for x in range(len(validationData)):
    v[x][validationTarget[x]] = 1

testingData = USPSMat
testingTarget = USPSTar
te = np.zeros(shape=(19999,10))
for x in range(len(testingData)):
    te[x][testingTarget[x]] = 1 
```

```python

def calculateSoftmax(x): # formula for softmax
  return (np.exp(x)/np.sum(np.exp(x)))

def prediction(data,weights):      #prediction 
    return(calculateSoftmax(np.dot(data,weights)))

def calculateErrorFunction(weights,target,data):     #caluclate the lost function
    predict = prediction(data,weights)
    error = (-1/len(target))*np.sum(np.multiply(target,np.log(predict)))
    return error


def probability(data,weights):         #return the highest probability 
    return np.argmax(calculateSoftmax((data.T*weights)),axis=1)

def getAccuracy(data,target,weights):     #compares the prediction with actually target value and return the accuracy
    weights = np.matrix(weights)
    final_predict = probability(data.T,weights)
    
    g = np.zeros(shape=(len(data), 10))
    for x in range(len(final_predict)):
        g[x][final_predict[x]] = 1
    correct = 0
    for x in range(0,len(g)):
        if(np.array_equal(g[x],target[x])):
            correct+=1
    accuracy = (correct/len(target))*100
    #correct = sum(g==target)
    #correct = correct/len(target)
    
    #return (correct*100)
    return accuracy
```

```python
W_Now = np.random.randn(trainingData.shape[1],10)*0.05

errorList = []
Val = []
TR = []
Test = []
TrainingAcc = []
ValidationAcc = []
TestingAcc = []
learningRate =0.004
index = 0
```

```python
for i in range(300):
    predict = prediction(trainingData,W_Now)
    error = calculateErrorFunction(W_Now, t, trainingData)
    print(error)
    errorList.append(error)
    gradient = np.dot(trainingData.T,predict-t)
    gradient = gradient/len(t)
    W_NEXT = W_Now - (learningRate * gradient)
    W_Now = W_NEXT
    TR = getAccuracy(trainingData,t, W_Now)
    TR = np.round(np.max(TR),3)
    print(TR)
    TrainingAcc.append(TR)
    Val = getAccuracy(validationData,v , W_Now)
    Val = np.round(np.max(Val),3)
    print(Val)
    ValidationAcc.append(Val)
    Test = getAccuracy(testingData, te, W_Now)
    Test = np.round(np.max(Test),3)
    print(Test)
    TestingAcc.append(Test)
print("Max Training Acc  = " + str(np.around(max(TrainingAcc), 5)))
print("Max Validation Acc = " + str(np.around(max(ValidationAcc), 5)))
print("Max Testing Acc    = " + str(np.around(max(TestingAcc), 5)))
print("cost = " + str(np.around(min(errorList), 5)))

Data.close()
```
